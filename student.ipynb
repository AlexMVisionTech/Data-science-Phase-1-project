### **Data Cleaning and Preprocessing**

**Dealing with the missing data**

```python
#Finding the null in the columns and rows to help with data cleaning
df_aviation.isnull().sum()
```

**Filling the location and Country with unkonwn because it requires a name string which needs accuracy**

**Remove the duplicates from the  `df_aviation` Dataframe**

```python
df_aviation = df_aviation.drop_duplicates()
```

**Filling the location and country with unknown because they require a name string**

```python
#Location
df_aviation['Location'] = df_aviation['Location'].fillna('Unknown')
#Country
df_aviation['Country'] = df_aviation['Country'].fillna('Unknown')
```

**Filling the null latitude and longitude with `0`**

```python
df_aviation['Latitude'] = df_aviation['Latitude'].fillna(0)
df_aviation['Longitude'] = df_aviation['Longitude'].fillna(0)
```

**Filling the `Airport.code` with the `unknown`**

```python
df_aviation['Airport.Code'] = df_aviation['Airport.Code'].fillna('Unknown')
```

**Using mode to fill the null in `Injury.Severity`**

```python
df_aviation['Injury.Severity'] = df_aviation['Injury.Severity'].fillna(df_aviation['Injury.Severity'].mode()[0])
```

**Use of foward filling in weather condition**

```python
#Using the foward filling to fill the empty cells in weather.condition
df_aviation['Weather.Condition'] = df_aviation['Weather.Condition'].fillna(method='ffill') 
```

**Using the `unkown` in the cells that prediction can be done due to its crucial need of the exact value eg the Aircraft Make**

```python
df_aviation['Make']=df_aviation['Make'].fillna('Unknown')
df_aviation['Aircraft.Category']=df_aviation['Aircraft.Category'].fillna('Unknown')
```

```python
df_aviation['Airport.Name']=df_aviation['Airport.Name'].fillna('Unknown')
```

**Use of `mean` to fill the cells that maybe filled with probability eg the data dealing with numbers**

```python
df_aviation["Total.Uninjured"] =df_aviation['Total.Uninjured'].fillna(df_aviation['Total.Uninjured'].mean())
```

```python
df_aviation["Total.Serious.Injuries"] =df_aviation['Total.Serious.Injuries'].fillna(df_aviation['Total.Serious.Injuries'].mean())
```

```python
df_aviation["Total.Fatal.Injuries"] =df_aviation['Total.Fatal.Injuries'].fillna(df_aviation['Total.Fatal.Injuries'].mean())
```

```python
df_aviation['Total.Minor.Injuries'] =df_aviation['Total.Minor.Injuries'].fillna(df_aviation['Total.Minor.Injuries'].mean())
```

```python
df_aviation['Model'] = df_aviation['Model'].fillna('Unknown')
```

#### **Saving of the already filtered dataframe as a csv file so that to easly use it in tableau for visualization**

```python
df_aviation.to_csv('updated_aviation_dataset.csv', index=False)
```

##                                                                           **Visualization**

**Coverting the data type of `Total.Fatal.Injuries` and `Total.uninjured` to int** 

```python
df_aviation['Total.Fatal.Injuries'] = df_aviation['Total.Fatal.Injuries'].astype(int)
df_aviation['Total.Uninjured'] = df_aviation['Total.Uninjured'].astype(int)
```

### **1. `Safety First`:Which aircraft models have the low fatality rate and high uninjured survival rate?**
          Choosing the aircraft makes with low fatality rates and high uninjured survival rates 

**A graph of `Fatal Injuries` vs `Uninjured` in Aviation Accidents**

```python
#Sizing of the graph
plt.figure(figsize=(14, 8))

#Plotting the Total.Fatal.Injuries with Total.Uninjured
plt.plot(df_aviation.index, df_aviation['Total.Fatal.Injuries'], label='Fatal Injuries', color='red', marker='o')
plt.plot(df_aviation.index, df_aviation['Total.Uninjured'], label='Uninjured', color='green', marker='o')

#Setting the labels 
plt.title('Fatal Injuries vs Uninjured in Aviation Accidents')
plt.xlabel('Index')
plt.ylabel('Count')
plt.legend()
plt.grid(True)

plt.show()
```

**A graph showing the relationship between the top 50 fatal injuries and Aircraft make**

```python
# Grouping by make then finding the total sum for the  total fatal injuries of each make 
top_fatal_injuries = df_aviation.groupby('Make')['Total.Fatal.Injuries'].sum().reset_index()

# Sorting the data in descending order based on 'Total.Fatal.Injuries'
top_fatal_injuries_sorted = top_fatal_injuries.sort_values(by='Total.Fatal.Injuries', ascending=False)

top_50_fatal_injuries = top_fatal_injuries_sorted.head(50)
```

```python
# Plotting the top 50 aircraft makes with the highest fatal injuries
plt.figure(figsize=(15, 6))
plt.bar(top_50_fatal_injuries['Make'], top_50_fatal_injuries['Total.Fatal.Injuries'], color='red')

# Adding  labels
plt.title('Top 50 Aircraft Makes with Highest Fatal Injuries')
plt.xlabel('Aircraft Make')
plt.ylabel('Total Fatal Injuries')
plt.xticks(rotation=90)  # Rotating x-axis labels for readability

plt.show()
```

**A Bar Graph showing the relationship between Top 50 Total.Uninjured and Make**

```python
# Groupping by 'Make' and summing the Total.Fatal.Injuries for each make
top_uninjured = df_aviation.groupby('Make')['Total.Uninjured'].sum().reset_index()

# Sortting the data in descending order as per Total.Fatal.Injuries
top_uninjured_sorted = top_uninjured.sort_values(by='Total.Uninjured', ascending=False)

# Get the top 50 aircraft makes with the highest fatal injuries
top_50_uninjured = top_uninjured_sorted.head(50)
```

```python
# Plotting the top 50 aircraft makes with the highest fatal injuries
plt.figure(figsize=(12, 6))
plt.bar(top_50_uninjured['Make'], top_50_uninjured['Total.Uninjured'], color='green')

# Addding labels
plt.title('Top 50 Aircraft Makes with Highest Uninjured')
plt.xlabel('Aircraft Make')
plt.ylabel('Total Uninjured')
plt.xticks(rotation=90)  

plt.show()
```

**A line graph showing the compination of `Total.Fatal.Injuries` and `Total.Uninjured` vs `Aircraft Make`**

```python
# Plotting the top 50 aircraft makes with the highest fatal injuries and uninjured counts
plt.figure(figsize=(12, 6))

# Plotting Fatal Injuries
plt.plot(top_50_fatal_injuries['Make'], top_50_fatal_injuries['Total.Fatal.Injuries'], marker='o', linestyle='-', color='red', label='Fatal Injuries')

# Plotting Uninjured counts
plt.plot(top_50_uninjured['Make'], top_50_uninjured['Total.Uninjured'], marker='s', linestyle='-', color='green', label='Uninjured')

# Add labels
plt.title('Comparison of Fatal Injuries and Uninjured Counts by Aircraft Make')
plt.xlabel('Aircraft Make')
plt.ylabel('Count')
plt.xticks(rotation=90)  
plt.legend()  

plt.show()
```
